### Assesment of Missingness

When looking at the "MONTH" column of the dataframe outages, we can see that there are several missing values throughout. The correlation between earlier years and missing data could be explained by impercise data recording methods, which have become more reliable in recent years due to better technology and training. We will conduct a permutation test to see if the missingness of the Month column is indeed missing at random dependent on the year column. 

When we initially look at the distribution of the Year column when the month is missing/not missing, we can see how there seems to be a large difference in the distributions for 3 specific years, 2000, 2002 and 2006. 

See Appendix Figure #7

We then compute the total variation distance of this distribution, and use this as our test statistic. When running a permutation test with 1000 simulations, we get a p-value of 0. This shows that the distributions are not similar when month is mssing vs. when it is not, thus Month is most likely dependent on the Year column. Since there are missing month values dependent on specific years, we are choosing to look at Years for our tests in regards to our question. Since our question is related to the election time in the US, rather than looking at months when elections are most prevalant (October, November, etc.), we are choosing to look at the election year as a whole. This way the missing month values do not interfere with our findings, and we can get a better picture of the differences in blackout durations between when there is and is not an election in America.

When further investigating the missingness of the month column, we test the dependency of the month column on the cause category column of the dataframe. ### Hypothesis Test
The null hypothesis for our project is as follows:
There is no difference in average duration in power outage between election years in swing states and non-election years in swing states, any deviation can be described by randomness.

The alternative hypothesis for our project is as follows:
There is a difference in average duration in power outage between election years in swing states and non-election years.

To test both of these hypotheses the difference of two means method was chosen because outage duration is a quantitative value that can be best described by its mean. The average outage duration across the fifteen complete years in each swing state was calculated to be 3482.45(2dp) minutes. In other words the from the start of 2000 to the end of 2015 the average length of a blackout across the swing states was 3482 minutes. This contrasted greatly to the average length of blackouts on election years which had an average of 4048.46(2dp) minutes, an almost 600 minute increase. To test whether this difference was due to chance or an alternative factor, a hypothesis test with a difference of two means method was used, which yielded a p-value of approximately 0.75, well above our significance level of 0.05. This supported our null hypothesis and rejected our alternative hypothesis.

See Appendix Figure #11

### Assessment of Missingness

```python
#Month Missingness and Year
# FIGURE 7
distr = (df.assign(is_null=df.MONTH.isnull()).pivot_table(index='is_null', columns='YEAR', aggfunc='size').fillna(0).apply(lambda x:x / x.sum(), axis=1)).T
distr.plot(kind = 'bar', title = "Distribution of Year when Month is null/not not")
```

```python
#Permutation test 1

tvds = []
shuffled = df.copy()
n = 1000

#Function for TVD
def tvd_of_groups(df):
        distr = (df.assign(is_null=df.MONTH.isnull())
                 .pivot_table(index='is_null', columns='YEAR', aggfunc='size').fillna(0)
                 .apply(lambda x:x / x.sum(), axis=1)).T
        tvd = np.sum(np.abs(distr[0] - distr[1])) / 2
        return tvd

#Calculating p-valuse 
observed = tvd_of_groups(df) 
for _ in range(n):
    s = df['YEAR'].sample(replace=False, frac=1).reset_index(drop=True)
    shuffled['YEAR'] = s
    tvds.append(tvd_of_groups(shuffled))
    
pval = (tvds >= observed).sum() / n
pval
```

###  Hypothesis Test

Now that we have cleaned and explored the data it's time to run a hypothesis test to explore the probability that there is a correlation between blackouts in swing states and election years.

```python
#Filter By Swing States

swing_states = ["Wisconsin", "Pennsylvania", "New Hampshire", "Minnesota",
                "Arizona", "Georgia", "Virginia", "Florida", "Michigan", 
                "Nevada", "Colorado", "North Carolina", "Maine"]

df_swing = df[df["U.S._STATE"].isin(swing_states)]
```

```python
#Mean Duration of Blackout across all swing states
#Drop 2016 as it is an incompleate year in the data set

df_swing = df_swing[df_swing["YEAR"] < 2016]
mean_blackout = np.mean(df_swing["OUTAGE.DURATION"])
mean_blackout
```

```python
# Average Blackout Duration for Each Election Year

#2000
df_2000 = df_swing.loc[df["YEAR"] == 2000]
mean_2000 = np.mean(df_2000["OUTAGE.DURATION"])
mean_2000
#2004
df_2004 = df_swing.loc[df["YEAR"] == 2004]
mean_2004 = np.mean(df_2004["OUTAGE.DURATION"])
mean_2004
#2008
df_2008 = df_swing.loc[df["YEAR"] == 2008]
mean_2008 = np.mean(df_2008["OUTAGE.DURATION"])
mean_2008
#2008
df_2012 = df_swing.loc[df["YEAR"] == 2012]
mean_2012 = np.mean(df_2012["OUTAGE.DURATION"])
mean_2012
{"2000":mean_2000, "2004":mean_2004, "2008":mean_2008, "2012":mean_2012, "Data Set Average" : mean_blackout}
```

```python
#Election Year Average
election_avg = (mean_2000 + mean_2004 + mean_2008 + mean_2012)/4
election_avg
```

```python
#Hypothesis Test
observed = abs(mean_blackout - election_avg)
election_years = [2000,2004,2008,2012]
sample_size= df_swing[df_swing["YEAR"].isin(election_years)].shape[0]

#Number of iterations to run
n = 1000


# Sample From Data Set
averages = []
for i in np.arange(n):
    random_sample = df.sample(int(sample_size), replace = False)
    new_avg = mean_blackout - np.mean(random_sample["OUTAGE.DURATION"])
    averages.append(abs(new_avg))

    
# Calculate P-Value
p_val = (np.array(averages) >= observed).sum() / n


{"P-val":p_val, "Swing State Average Blackout On Election Years":election_avg, 
 "Swing State Average Blackout Across All Years" : mean_blackout,
"Observed Value": observed}
```
